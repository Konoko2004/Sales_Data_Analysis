```python
# Linear Regression
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

height = [[1.6],[1.65],[1.7],[1.73],[1.8]]
weight = [[60],[65],[72.3],[75],[80]]

plt.plot(height,weight,'k.')

# Create and fit the model
model = LinearRegression()
model.fit(X = height, y = weight)

#make prediction 
weight = model.predict([[1.75]])[0][0]
print(round(weight,2))
```

```python
# Linear Regression line

%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

height = [[1.6],[1.65],[1.7],[1.73],[1.8]]
weight = [[60],[65],[72.3],[75],[80]]

plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters.')
plt.ylabel('Weights in Kilograms.')
plt.plot(height,weight,'k.')
plt.axis([1.5,1.85,50,90])
plt.grid(True)

#Plot the regression line
plt.plot(height,model.predict(height),color='r')
```

```python
# Linear Regression - Getting the Gradient and the intercept of the Linear Regression Line

plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters.')
plt.ylabel('Weights in Kilograms.')
plt.plot(height,weight,'k.')

#we set the x axis to start from 0 and then enlarge the y axis.

plt.axis([0,1.85,-200,200])
plt.grid(True)

#plot the regression line :

extreme_height = [[0],[1.8]]
plt.plot(extreme_height, model.predict(extreme_height),color='r')

#What is the value of y when x = 0
predicted_Y = model.predict([[0]])
predicted_Y = round(predicted_Y[0][0],2)
print(predicted_Y)

#Alternatively we can use .intercept property

print(round(model.intercept_[0],2))

#what is the gradient of the linear regression

print(round(model.coef_[0][0],2))
```

```python
#Examining the Performance of the Model by Calculating the Residual Sum of Squares

import numpy as np
print('Residual sum of squares: %.2f' % np.sum((weight - model.predict(height))**2))     #model.predict takes in a 2d array. it returns the values of Y along the regression line based on the values of the x axis

```

#Evaluating the model using a Test Dataset

```python
#Finding the R-Squared - 
#Step 1 : Find TSS
#Step 2 : Find RSS
#Step 3 : R-Squared = 1 - RSS / TSS

#test data

height_test = [[1.58],[1.62],[1.69],[1.76],[1.82]]
weight_test = [[58],[63],[72],[73],[85]]



#Total Sum of Squares (TSS)

weight_test_mean = np.mean(np.ravel(weight_test))

#TSS formula :
TSS = np.sum((np.ravel(weight_test) - weight_test_mean) ** 2)
print("TSS : %.2f" % TSS)


#Residual Sum of Squares (RSS)

#RSS formula :
RSS = np.sum((np.ravel(weight_test) - 
              np.ravel(model.predict(height_test)))**2)

print('RSS : %.2f' % RSS)


#R-Squared
R_Squared = 1 - RSS / TSS
print("R-Squared : %.2f" % R_Squared)
```

```python
#Short cut : using the model.score function (This function basically does the entire code in the above cell)

#using scikit-learn to calculate r-squared
height_test = [[1.58],[1.62],[1.69],[1.76],[1.82]]
weight_test = [[58],[63],[72],[73],[85]]
print('R-squared :%.2f' % model.score(height_test,weight_test))
```

# Persisting the Model

```python
import pickle

#save the model to disk
filename = 'HeightsAndWeights_model.sav'
#write to the file using write and binary mode
pickle.dump(model,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open(filename,'rb'))

result = loaded_model.score(height_test,weight_test)

print(result) #result is r-squared
```
