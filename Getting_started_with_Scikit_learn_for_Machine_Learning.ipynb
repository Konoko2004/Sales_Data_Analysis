```python
# Linearly distributed dataset

%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_regression

x,y = make_regression(n_samples = 100,n_features = 1, noise =5.4)
plt.scatter(x,y)
```

```python
# Clustered Dataset

%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples = 500, centers = 3)
rgb = np.array(['r','g','b'])

#plot the blobs using scatter plot and use color coding

plt.scatter(x[:,0],x[:,1],c= rgb[y])
```
