```python
# Linearly distributed dataset

%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_regression

x,y = make_regression(n_samples = 100,n_features = 1, noise =5.4)
plt.scatter(x,y)
```

```python
# Clustered Dataset

%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples = 500, centers = 3)
rgb = np.array(['r','g','b'])

#plot the blobs using scatter plot and use color coding

plt.scatter(x[:,0],x[:,1],c= rgb[y])
```

```python
# Linear Regression
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

height = [[1.6],[1.65],[1.7],[1.73],[1.8]]
weight = [[60],[65],[72.3],[75],[80]]

plt.plot(height,weight,'k.')

# Create and fit the model
model = LinearRegression()
model.fit(X = height, y = weight)

#make prediction 
weight = model.predict([[1.75]])[0][0]
print(round(weight,2))
```

```python
# Linear Regression line

%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

height = [[1.6],[1.65],[1.7],[1.73],[1.8]]
weight = [[60],[65],[72.3],[75],[80]]

plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters.')
plt.ylabel('Weights in Kilograms.')
plt.plot(height,weight,'k.')
plt.axis([1.5,1.85,50,90])
plt.grid(True)

#Plot the regression line
plt.plot(height,model.predict(height),color='r')
```

```python
# Linear Regression - Getting the Gradient and the intercept of the Linear Regression Line

plt.title('Weights plotted against heights')
plt.xlabel('Heights in meters.')
plt.ylabel('Weights in Kilograms.')
plt.plot(height,weight,'k.')

#we set the x axis to start from 0 and then enlarge the y axis.

plt.axis([0,1.85,-200,200])
plt.grid(True)

#plot the regression line :

extreme_height = [[0],[1.8]]
plt.plot(extreme_height, model.predict(extreme_height),color='r')

#What is the value of y when x = 0
predicted_Y = model.predict([[0]])
predicted_Y = round(predicted_Y[0][0],2)
print(predicted_Y)

#Alternatively we can use .intercept property

print(round(model.intercept_[0],2))

#what is the gradient of the linear regression

print(round(model.coef_[0][0],2))
```
