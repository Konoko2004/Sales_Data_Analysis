```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

raw_data = pd.read_csv(r'https://raw.githubusercontent.com/Konoko2004/Sales_Data_Analysis/main/sales_data.csv')
raw_data.head()
```

```python
a =raw_data.groupby('Date')['Profit'].sum()

a.sort_values(inplace = True)
a = pd.DataFrame(a)
```

```python
b = pd.DataFrame(a).quantile(q=[0, .25, .50, .75, 1])
b
```

```python
np.percentile(a, [0, 25, 50, 75, 100])
```

```python
a['Percentile'] = pd.qcut(a['Profit'],q = [0,0.25,0.5,0.75,1], labels = ['25th pct','50th pct','75 pct','100 pct'])
a
```
