```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston
dataset = load_boston()
```

```python
# print(dataset.feature_names)
# print(dataset.data)
# print(dataset.DESCR)
print(dataset.target)
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
df['MEDV'] = dataset.target
df.head()
df.corr()
```

```python
# print(df.corr().abs().nlargest(3,'MEDV').index)          #we need abs to remove negatives. We want to find the top 2 correlated features with our target.
print(df.corr().abs().nlargest(3,'MEDV').values[:,13])
```

```python
#relationship between LSTAT and MEDV

plt.scatter(df['LSTAT'],df['MEDV'],marker='o')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
```

```python
#relationship between RM and MEDV

plt.scatter(df['RM'],df['MEDV'],marker='o')
plt.xlabel('RM')
plt.ylabel('MEDV')
```
